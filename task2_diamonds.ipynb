## Section 1-Load: Get the file, store it in your repo, and load it into a DataFrame. 

```python
import pandas as pd # import pandas library
import numpy as np # import numpy library

diamondsdf = pd.read_csv('diamonds.csv', index_col=0) # Create a DataFrame from csv file

# renaming other columns with small changes for better understanding measures
diamondsdf.columns = ['carat', 'cut', 'color', 'clarity', 'depth(%)', 'table(%)', 'price(USD)', 'X(mm)', 'Y(mm)', 'Z(mm)'] 

diamondsdf # call DataFrame
```

## Section 5-Unique: Use the Series method unique() to get unique category values. 

```python
unique_cut = diamondsdf['cut'].unique() #Quality in increasing order Fair, Good, Very Good, Premium, Ideal
print(unique_cut)
```

```python
unique_clarity = diamondsdf['clarity'].unique() #How obvious inclusions are within the diamond:(in order from best to worst, FL = flawless, I3= level 3 inclusions) FL,IF, VVS1,
print(unique_clarity)
```

```python
unique_color = diamondsdf['color'].unique() #Color of the diamond, with D being the best and J the worst
print(unique_color)
```

## Section 6-Histograms: Use the DataFrame's hist() function to create a histogram for each numerical column.

```python
import matplotlib.pyplot as plt

#enable Matplotlib support
%matplotlib inline

columns = ['carat', 'depth(%)', 'table(%)', 'price(USD)', 'X(mm)', 'Y(mm)', 'Z(mm)'] # make list of columns for loop

for col in columns: #loop through each column making a histogram
    histogram = diamondsdf.hist(col) # Produce histogram
    plt.xlabel(col.title()) #label X axis
    plt.title(f'Histogram of {col.title()}') # set title
    plt.ylabel('Frequency') #label Y axis



```
